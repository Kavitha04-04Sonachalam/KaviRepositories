4:NUMPY PROGRAM TO FIND REAL AND IMAGINERY PARTS OF ARRAY

```python
import numpy as np

# Create an array of complex numbers
complex_numbers = np.array([1 + 2j, 3 - 4j, 5 + 6j, 7 - 8j])

# Extract real and imaginary parts
real_parts = np.real(complex_numbers)
imaginary_parts = np.imag(complex_numbers)

```

```python
# Display the results
print("Original Complex Numbers:")
print(complex_numbers)
print("\nReal Parts:")
print(real_parts)
print("\nImaginary Parts:")
print(imaginary_parts)
```
